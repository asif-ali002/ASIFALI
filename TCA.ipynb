```python
import pandas as pd
import matplotlib.pyplot as plt

def stacked_percent_bar(df, x, hue):
    # Crosstab counts
    ct = pd.crosstab(df[x], df[hue])

    # Convert to row percentages (each bar = 100%)
    ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

    # Choose fixed colors for hue levels
    colors = ["blue", "orange"]  # Adjust if hue has more than 2 categories

    # Plot
    ax = ct_percent.plot(kind="bar", stacked=True, figsize=(7,5), color=colors)

    # Add % labels
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x_pos, y_pos = p.get_xy()
        if height > 0:
            ax.text(x_pos + width/2, y_pos + height/2,
                    f'{height:.1f}%', ha="center", va="center",
                    fontsize=10, color="white", fontweight="bold")

    plt.title(f"{hue} by {x} (% within each {x})")
    plt.ylabel("Percentage")
    plt.xlabel(x)
    plt.legend(title=hue)
    plt.show()


# Example usage:
stacked_percent_bar(df, x="SeniorCitizen", hue="Churn")
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)




# List of categorical columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define subplot grid
n_rows = 3
n_cols = 3
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 12))

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through columns and create countplots
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i], hue = "Churn")
    axes[i].set_title(f"Countplot of {col}", fontsize=12)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

# Hide any unused subplots (if cols < grid size)
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
